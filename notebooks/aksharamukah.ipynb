```python
import pandas as pd
from aksharamukha import transliterate
```

```python


# Latin Pali text (using standard IAST/ISO diacritics)
pali_text = "Namo tassa bhagavato arahato sammāsambuddhassa"

# Convert from Roman (IAST) to Sinhala
# 'IAST' covers the diacritics like ā, ṃ, ṇ, etc.
sinhala_text = transliterate.process('IAST', 'Sinhala', pali_text)

print(sinhala_text)
```

```python
df = pd.read_csv("pali_canon.csv", index_col='index')
```

```python
dn = df[df['nikaya'] == 'dn']
dn['hela_text'] = dn['text'].apply(lambda x: transliterate.process('IAST', 'Sinhala', x))
dn
```

```python
mn = df[df['nikaya'] == 'mn']
mn['hela_text'] = mn['text'].apply(lambda x: transliterate.process('IAST', 'Sinhala', x))
mn
```

```python
kn = df[df['nikaya'] == 'kn']
kn['hela_text'] = kn['text'].apply(lambda x: transliterate.process('IAST', 'Sinhala', x))
kn
```

```python
sn = df[df['nikaya'] == 'sn']

sn['hela_text'] = sn['text'].apply(lambda x: transliterate.process('IAST', 'Sinhala', x) if pd.notna(x) else x)

sn
```

```python
an = df[df['nikaya'] == 'an']

an['hela_text'] = an['text'].apply(lambda x: transliterate.process('IAST', 'Sinhala', x) if pd.notna(x) else x)

an
```

```python
df_hela = pd.concat([dn, mn, kn, sn, an])

df_hela.to_csv("pali_canon_hela.csv", index=True, index_label='index')
```
