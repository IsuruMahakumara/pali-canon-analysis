```python
from arango import ArangoClient

# 1. Initialize the ArangoDB client
# Assuming it's running locally; replace 'localhost' with your server IP if remote
client = ArangoClient(hosts='http://localhost:8529')

# 2. Connect to the "_system" database
# Replace 'password' with your actual root password
db = client.db('_system', username='root', password='test')
```

```python
db_suttacentral = client.db('suttacentral', username='root', password='test')
```

```python
collections = db_suttacentral.collections()

for col in collections:
    # We filter for non-system collections (those that don't start with '_')
    if not col['system']:
        print(f"- {col['name']} (Type: {col['type']})")
```

```python
col = db.collection('fallen_leaves')

# Fetch statistics
stats = col.statistics()

# 1. Get the number of documents
doc_count = stats['count']

# 2. Get the size in bytes (approximate physical size)
# Note: 'size' refers to the total allocated space for data
total_size_bytes = stats['figures']['datafiles']['fileSize'] 

print(f"Collection: {col.name}")
print(f"Document Count: {doc_count}")
print(f"Disk Usage: {total_size_bytes / 1024 / 1024:.2f} MB")
```

```python
# This is the most "honest" way to check if it's there and has data
query = "RETURN LENGTH(@@col)"
try:
    count = db.aql.execute(query, bind_vars={'@col': 'sc_bilara_texts'}).next()
    print(f"Success! Collection exists and contains {count} documents.")
except Exception as e:
    print(f"AQL failed: {e}")
```
