### Cleaning  Complex dicts

```python
dict_path = '../dictionaries/complex/en/pli2en_pts.json'

with open(dict_path, 'r') as f:
    dict_data = json.load(f)



html = dict_data[1]['text']
```

```python
soup = BeautifulSoup(html, "html.parser")

# Extract ordered list items
ol_items = [li.get_text(strip=True, separator=" ") for li in soup.select("dd#Aṁsa-1 ol li")]

# Extract unordered list items
ul_items = [li.get_text(strip=True, separator=" ") for li in soup.select("dd#Aṁsa-1 ul.compounds li")]

# Combine if you want
all_items = ol_items + ul_items

all_items
```

```python
ul_items
```

```python
dl_dict = {}

# Find the <dl>
dl = soup.find("dl")

# Iterate over all children of <dl>
for child in dl.children:
    if child.name == "dt":
        # Extract term text
        term = child.get_text(strip=True, separator=" ")
    elif child.name == "dd":
        # Extract definition text
        definition = child.get_text(strip=True, separator=" ")
        dl_dict[term] = definition

# Result
from pprint import pprint
pprint(dl_dict)
```

```python

def extract_li_as_dict(li_tag):
    """Convert a <li> into a dict with 'text' and optional 'subitems'."""
    text_parts = []
    subitems_dict = {}

    for child in li_tag.children:
        if child.name in ["ol", "ul"]:
            # Convert nested list into dict using index-based keys
            for idx, subli in enumerate(child.find_all("li", recursive=False), start=1):
                subitems_dict[f"item_{idx}"] = extract_li_as_dict(subli)
        elif isinstance(child, str) and child.strip():
            text_parts.append(child.strip())
        elif child.name:
            # Include text from other tags inside li
            text = child.get_text(strip=True, separator=" ")
            if text:
                text_parts.append(text)

    text = " ".join(text_parts).strip()
    if subitems_dict:
        return {"text": text, "subitems": subitems_dict}
    else:
        return {"text": text}

def extract_list_as_dict(list_tag):
    """Convert <ol> or <ul> into dict with numbered keys."""
    result = {}
    for idx, li in enumerate(list_tag.find_all("li", recursive=False), start=1):
        result[f"item_{idx}"] = extract_li_as_dict(li)
    return result

# Build the DL dictionary
dl_dict = {}
dl = soup.find("dl")

current_dt = None
for child in dl.children:
    if child.name == "dt":
        current_dt = child.get_text(strip=True, separator=" ")
    elif child.name == "dd" and current_dt:
        dd_dict = {}
        for subchild in child.children:
            if subchild.name in ["ol", "ul"]:
                dd_dict.update(extract_list_as_dict(subchild))
            elif subchild.name:
                text = subchild.get_text(strip=True, separator=" ")
                if text:
                    dd_dict["text"] = dd_dict.get("text", "") + (" " if "text" in dd_dict else "") + text
            elif isinstance(subchild, str) and subchild.strip():
                dd_dict["text"] = dd_dict.get("text", "") + (" " if "text" in dd_dict else "") + subchild.strip()
        dl_dict[current_dt] = dd_dict

from pprint import pprint
pprint(dl_dict)
```
